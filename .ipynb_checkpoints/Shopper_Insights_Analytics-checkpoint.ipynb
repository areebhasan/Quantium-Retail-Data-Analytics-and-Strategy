## **Introduction**

This notebook presents a comprehensive analysis of customer purchasing behavior within the retail sector, focusing on the chip category at Woolworths supermarket. The objective is to derive actionable insights that can inform the supermarketâ€™s strategic planning, particularly in the areas of marketing, product placement, and inventory management.

To achieve this, the analysis is structured as follows:

1. **Data Preparation:** 
   - Importing and cleaning transaction and customer data.
   - Ensuring the data is ready for analysis by handling missing values, correcting data types, and removing irrelevant entries.

2. **Exploratory Data Analysis (EDA):**

5. **Brand Affinity Analysis:**

### T-Testing
##### --> We are comparing the total sales (TOT_SALES) between two groups of customers: "Mainstream" and "Premium" customers who are in the "Young Singles/Couples" life stage. We want to see if the average sales differ significantly between these two groups.

```python
from scipy.stats import ttest_ind

# Filtering data for Mainstream and Premium customers in the Young Singles/Couples segment
mainstream = merged_data[(merged_data['PREMIUM_CUSTOMER'] == 'Mainstream') & (merged_data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES')]
premium = merged_data[(merged_data['PREMIUM_CUSTOMER'] == 'Premium') & (merged_data['LIFESTAGE'] == 'YOUNG SINGLES/COUPLES')]

# Calculating the mean TOT_SALES for both segments
mainstream_mean = mainstream['TOT_SALES'].mean()
premium_mean = premium['TOT_SALES'].mean()

# Printing the means
print(f"Mean TOT_SALES for Mainstream customers: {mainstream_mean}")
print(f"Mean TOT_SALES for Premium customers: {premium_mean}")

# Performing the T-test
t_stat, p_value = ttest_ind(mainstream['TOT_SALES'], premium['TOT_SALES'])

# Printing the T-test results
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

#### Observation on T-Test for TOT_SALES between Mainstream and Premium Customers

The analysis compared the mean total sales (`TOT_SALES`) between "Mainstream" and "Premium" customers within the "Young Singles/Couples" segment.

- **Mean TOT_SALES**:
  - **Mainstream customers**: 7.53
  - **Premium customers**: 6.65

The results show that "Mainstream" customers have a higher average total sales compared to "Premium" customers.

- **T-test Results**:
  - **T-statistic**: 23.58
  - **P-value**: 1.21e-121

Given the extremely low P-value, which is much smaller than any conventional significance level (e.g., 0.05), we reject the null hypothesis that there is no difference in mean TOT_SALES between the two groups. This suggests a statistically significant difference in total sales between "Mainstream" and "Premium" customers in this segment, with "Mainstream" customers spending more on average.

#### Analysing Brand Affinity for Mainstream, Young Singles/Couples

This analysis compares brand affinity between the "Mainstream, Young Singles/Couples" segment and the rest of the population.

##### Steps:
1. **Segment Filtering**: Separating data for the target segment and others from `merged_data`.
2. **Calculate Quantities**: Determining product quantities for each segment.
3. **Brand Proportions**: Computing brand proportions within each segment.
4. **Brand Affinity**: Merging and calculate brand affinity, then sort by affinity score.


```python
# Filter data for Mainstream, Young Singles/Couples segment
segment1 = merged_data[(merged_data['LIFESTAGE'] == "YOUNG SINGLES/COUPLES") & (merged_data['PREMIUM_CUSTOMER'] == "Mainstream")]

# Filter data for the rest of the population
other = merged_data[~((merged_data['LIFESTAGE'] == "YOUNG SINGLES/COUPLES") & (merged_data['PREMIUM_CUSTOMER'] == "Mainstream"))]

# Calculate total product quantity for segment1 and other
quantity_segment1 = segment1['PROD_QTY'].sum()
quantity_other = other['PROD_QTY'].sum()

# Calculate brand proportions for segment1 and other
quantity_segment1_by_brand = segment1.groupby('BRAND')['PROD_QTY'].sum().reset_index()
quantity_segment1_by_brand['targetSegment'] = quantity_segment1_by_brand['PROD_QTY'] / quantity_segment1

quantity_other_by_brand = other.groupby('BRAND')['PROD_QTY'].sum().reset_index()
quantity_other_by_brand['other'] = quantity_other_by_brand['PROD_QTY'] / quantity_other

# Merge and calculate brand affinity
brand_proportions = pd.merge(quantity_segment1_by_brand[['BRAND', 'targetSegment']], 
                             quantity_other_by_brand[['BRAND', 'other']], on='BRAND')
brand_proportions['affinityToBrand'] = brand_proportions['targetSegment'] / brand_proportions['other']

# Sort by affinity to brand
brand_proportions_sorted = brand_proportions.sort_values(by='affinityToBrand', ascending=False)
brand_proportions_sorted
```

#### Summary on Brand Affinity for "Mainstream, Young Singles/Couples" Segment

The analysis reveals that the "Mainstream, Young Singles/Couples" segment has the highest affinity for the following brands:

1. **TYRRELLS**: With an affinity score of 1.233, this brand is particularly favored by the target segment, indicating they purchase this brand 23.3% more than the rest of the population.
2. **TWISTIES**: Close behind, with an affinity score of 1.224, showing a strong preference by the segment.
3. **DORITOS, KETTLE, TOSTITOS, and PRINGLES**: These brands also have high affinity scores above 1.19, making them popular choices within this segment.

On the other hand, brands like **BURGER**, **WOOLWORTHS**, and **SUNBITES** have significantly lower affinity scores, indicating that these brands are less favored by the target segment compared to the general population.

This insight suggests that marketing efforts for the "Mainstream, Young Singles/Couples" segment could be more effective if focused on the top-affinity brands such as TYRRELLS, TWISTIES, and DORITOS.
